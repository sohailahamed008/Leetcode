```python
df2=df.copy()
```

```python
df2=df2.drop(df2.index[10472])
```

```python
df2[~df2['Reviews'].str.isnumeric()]
```

```python
## convert reviews datatype to int
df2['Reviews']=df2['Reviews'].astype(int)
```

```python
df2.info()
```

```python
df2['Size'].unique()
```

```python
df2['Size'].isnull().sum()
```

```python
df2['Size']=df2['Size'].str.replace('M','000')
df2['Size']=df2['Size'].str.replace('K','')
df2['Size']=df2['Size'].replace('Varies with device',np.nan)
df2['Size']=df2['Size'].astype(float)
```

```python
df2.info()
```

```python
df2.isnull().sum()
```

```python
df2['Installs'].unique()
```

```python
df2['Installs']=df2['Installs'].str.replace('+','')
df2['Installs']=df2['Installs'].str.replace(',','')
df2['Installs']=df2['Installs'].astype(int)
```

```python
df2["Price"].unique()
```

```python
df2['Price']=df2['Price'].str.replace('$','')
df2['Price']=df2['Price'].astype(float)
```

```python
df2["Price"].unique()
```

```python
df2["Price"].unique()
```

```python
df2.info()
```

```python
df2['Last Updated'].unique()
```

```python
df2['Last Updated']=pd.to_datetime(df2['Last Updated'])
```

```python
df2['Day']=pd.to_datetime(df2['Last Updated']).dt.day
df2['Month']=pd.to_datetime(df2['Last Updated']).dt.month
df2['Year']=pd.to_datetime(df2['Last Updated']).dt.year
```

```python
df2.info()
```

```python
df2.to_csv('Cleaned_gps.csv')
```
